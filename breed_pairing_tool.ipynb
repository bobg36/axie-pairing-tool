```python
virgin = 11892685

#remove incest pairs
virgin_details = fn.get_virgin_details(virgin)
virgin_hexgene = virgin_details[1]
virgin_sire = virgin_details[4]
virgin_matron = virgin_details[5]
no_incest_list = [] #no_incest_list is same as breed pool but siblings with same parents are removed
for compare_details in breed_pool:
    parent_list = [virgin_sire, virgin_matron]
    sire = compare_details[4]
    matron = compare_details[5]
    parent_list.append(sire)
    parent_list.append(matron)
    unique_count = len(set(parent_list))
    if unique_count == 4:
        no_incest_list.append(compare_details)

print('finding mate for: ' + str(virgin) + ', purity' + str(fn.get_purity(virgin_hexgene)))
print(virgin_details)
print(str(len(no_incest_list)) + ' axies left in the breeding pool after removing siblings')

#get c_score, purity, marketplace link
unsorted_matchings = []
for compare_details in no_incest_list:
    final_axie_details = []
    axieId = compare_details[0]
    axieClass = compare_details[2]
    mplink = 'https://app.axieinfinity.com/marketplace/axies/' + str(axieId)
    compare_hexgene = compare_details[1]
    purity = fn.get_purity(compare_hexgene)
    c_score = fn.get_c_score(virgin_hexgene, compare_hexgene)
    final_axie_details.append(c_score)
    final_axie_details.append(purity)
    final_axie_details.append(axieId)
    final_axie_details.append(mplink)
    final_axie_details.append(axieClass)
    unsorted_matchings.append(final_axie_details)

#sort and show breed pairings
matchings = sorted(unsorted_matchings, key=lambda x: x[0])
for item in matchings:
    print(item)
```
